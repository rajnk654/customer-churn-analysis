```python
# To load a CSV file using pandas in Python
import pandas as pd
df = pd.read_csv("../Customer_telco_churn.csv")
```

```python
#Displaying random sample of 10 rows from the DataFrame
#sample method is used to return a random number of rows from the dataframe
df.sample(10)
```

```python
# To check how many datatypes are present in each column of the DataFrame
df.info()
```

```python
#Drop unnecessary columns
#Dropping columns 'customerID' and 'Unnamed: 0' as they are not needed for analysis
df = df.drop(columns=['customerID'])
```

```python
# To verify that the columns have been dropped
df.info()
```

```python
# Analyze categorical features by printing the count of unique values,
# which helps in selecting appropriate encoding techniques.

numerical_features_list = ['tenure', 'MonthlyCharges', 'TotalCharges']
for col in df.columns:
    if col not in numerical_features_list:
        print(col, ":", df[col].nunique())
        print("-" * 75)
```

```python
#we see that TotalCharges has some issues because it should be numerical but has some non-numeric values. 
# So we need to convert it to numeric.
# First, we can replace the spaces with NaN or 0.0

df[df['TotalCharges']==" "]
```

```python
# So we need to convert it to numeric.
df['TotalCharges']=df['TotalCharges'].replace(" ",0.0)
```

```python
#we see that TotalCharges has some issues because it should be numerical but has some non-numeric values. 
# So we need to convert it to numeric.
df['TotalCharges']=df['TotalCharges'].astype(float)
```

```python
df.info()
```

```python
# Identifying columns with categorical data type
object_columns = df.select_dtypes(include=['object']).columns
print(object_columns)
```

```python
#Applying label encoding to binary categorical features
#We converting all categorical features with two unique values to numerical format using label encoding.
#Because many machine learning algorithms require numerical input.

#We are not using one-hot encoding here because it can lead to the "curse of dimensionality(Means dimensionality of the dataframe goes up very high)"
# when applied to features with many unique values.Therefore,label encoding is more efficient for binary categorical features.
from sklearn.preprocessing import LabelEncoder


#Initialize a dictionary to save label encoders for each binary feature
encoders ={}

#Apply label encoding store the encoders
for column in object_columns:
    label_encoders = LabelEncoder()
    df[column] = label_encoders.fit_transform(df[column])
    encoders[column] = label_encoders
    
    
```

```python
encoders
```

```python
# To check the dataframe after label encoding
df.head(20)
```

```python
#Seperating features and target variable
X = df.drop('Churn', axis=1)
y = df['Churn']
```

```python
print(X.shape)
```

```python
#  the columns to be scaled

col_to_scale =['tenure','MonthlyCharges','TotalCharges']
df_numeric  = df[col_to_scale]
```

```python
#Importing min-max scaler from sklearn library 

from sklearn.preprocessing import MinMaxScaler

#Applying min-max scaling to the numerical features
min_max_scaler = MinMaxScaler()
df_minmax = pd.DataFrame(min_max_scaler.fit_transform(df_numeric), columns=col_to_scale)

print("Min_Max Scalared Dataframe",df_minmax.head())
```

```python
# Import necessary libraries for model training and evaluation
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Train-Test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

# Decision Tree model
dt_model = DecisionTreeClassifier(
    max_depth=5,
    random_state=42
)

# Train the model
dt_model.fit(X_train, y_train)

# Predictions
y_pred = dt_model.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
