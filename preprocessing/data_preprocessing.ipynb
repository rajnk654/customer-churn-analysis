## Steps For Data Preprocessing Of Dataset

```python
# To load a CSV file using pandas in Python
import pandas as pd
df = pd.read_csv("../Customer_telco_churn.csv")
```

```python
# Display the first few rows of the dataframe
df.head()  
```

```python
# Basic information about the dataset
df.info()
```

```python
# drop unnecessary columns
#Like customerID column which is not useful for analysis
df = df.drop(columns=['customerID'])
```

```python
# To check customerID column is dropped
df.head()
```

```python
# Checking for missing values in the dataset
df.isnull().sum() # This method will give the count of missing values in each column
```

```python
# Analyze categorical features by printing the count of unique values,
# which helps in selecting appropriate encoding techniques.
numerical_features_list = ['tenure', 'MonthlyCharges', 'TotalCharges']
for col in df.columns:
    if col not in numerical_features_list:
        print(col, ":", df[col].nunique())
        print("-" * 50)
```

```python
#we see that TotalCharges has some issues because it should be numerical but has some non-numeric values. 
# So we need to convert it to numeric.
# First, we can replace the spaces with NaN or 0.0
df[df["TotalCharges"]==" "]
```

```python
# So we need to convert it to numeric.
df['TotalCharges']=df['TotalCharges'].replace(" ","0.0")
```

```python
#we see that TotalCharges has some issues because it should be numerical but has some non-numeric values. 
# So we need to convert it to numeric.
df['TotalCharges']= df["TotalCharges"].astype(float)
```

```python
# To confirm the changes are made
df.info() 
```
