```python
# To load a CSV file using pandas in Python
import pandas as pd
df = pd.read_csv("../Customer_telco_churn.csv")
```

```python
#Displaying random sample of 12 rows from the DataFrame
#sample method is used to return a random number of rows from the dataframe
df.sample(12)
```

```python
#Drop unnecessary colums
#Dropping columns 'customerID' and 'Unnamed: 0' as they are not needed for analysis
df = df.drop(columns=['customerID'])
```

```python
# To verify that the columns have been dropped
df.info()
```

```python
#To check we have any missing values in the dataset
df.isnull().sum()
```

```python
# Analyze categorical features by printing the count of unique values,
# which helps in selecting appropriate encoding techniques.

numerical_features_list = ['tenure', 'MonthlyCharges', 'TotalCharges']
for col in df.columns:
    if col not in numerical_features_list:
        print(col, ":", df[col].nunique())
        print("-" * 35)
```

```python
#we see that TotalCharges has some issues because it should be numerical but has some non-numeric values. 
# So we need to convert it to numeric.
# First, we can replace the spaces with NaN or 0.0

df[df['TotalCharges']==" "]
```

```python
# So we need to convert it to numeric.
df['TotalCharges']=df['TotalCharges'].replace(" ",0.0)
```

```python
#we see that TotalCharges has some issues because it should be numerical but has some non-numeric values. 
# So we need to convert it to numeric.
df['TotalCharges']=df['TotalCharges'].astype(float)
```

```python
#To verify the TotalCharges column datatype is converted to float
df.info()
```

```python
# Identifying columns with categorical data type
object_columns = df.select_dtypes(include=['object']).columns
print(object_columns)
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

#  Select categorical (object) columns
object_columns = df.select_dtypes(include="object").columns

#  Apply One-Hot Encoding
# drop='first' removes one category per column (this is normal)
encoder = OneHotEncoder(sparse_output=False)

encoded_array = encoder.fit_transform(df[object_columns])

#  Convert encoded array to DataFrame (0 and 1)
encoded_df = pd.DataFrame(
    encoded_array.astype(int),
    columns=encoder.get_feature_names_out(object_columns),
    index=df.index
)

#  Combine encoded columns with numerical columns
df_final = pd.concat([df.drop(columns=object_columns), encoded_df], axis=1)

#  Display output
print(df_final.head())

#  Check data types (should be int for encoded columns)
print(df_final.dtypes)
```
